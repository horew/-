## 绘制散点图 Scatter

```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 设置绘制风格
plt.style.use('ggplot')
# plt.style.use('seaborn-whitegrid')

# 设置画布大小
plt.rcParams["figure.figsize"] = (10,6)
```

```python
import matplotlib
matplotlib.rc('xtick', labelsize=16)
matplotlib.rc('ytick', labelsize=16)
```

### 1、使用 ``plt.plot`` 绘制散点图 Scatter 

- ``plt.plot``/``ax.plot``也可以画散点图:

```python
x = np.linspace(0, 10, 20)
y = np.sin(x)

plt.plot(x, y, 'd', markersize=5, color='black');
plt.axis('equal')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))

plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

参数写在一起：

```python
plt.plot(x, y,'-og', linewidth=3, markersize=8);
```

可以通过 ``plt.plot`` 设置一系列属性:

```python
plt.plot(x, y, '-h', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='red',
         markeredgewidth=1)
plt.ylim(-1.2, 1.2);
```

### 2、使用 ``plt.scatter`` 绘制散点图

```python
plt.scatter(x, y, marker='p', s=180);
```

#### ``plt.scatter`` 和 ``plt.plot`` 的最主要区别是前者可以控制每个 marker 的 (size, face color, edge color, etc.)

#### 下图使用 scatter 绘制随机大小和颜色的散点

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();
```

#### 参数 color 的数值自动转化为相应的颜色（具体根据 ``colorbar()`` 命令指定的参数）, 参数 size 指的是像素。

#### 可用于绘制多维数据。

#### 绘制 Iris 花卉数据

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.6, s=100*features[3], c=iris.target, cmap='viridis')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

```python
features.shape
```

#### 可以展示 4 维数据:
#### (x, y) 对应 sepal length and width，点的大小对应 petal width，颜色对应花卉的种类。

## ``plot`` vs ``scatter``: 绘制效率

``plt.plot`` 的绘制效率高于 ``plt.scatter``。

因此当数据量大的时候 ``plt.plot``更好，但需要针对数据绘制不同颜色和大小是可以使用 ``plt.scatter``。
